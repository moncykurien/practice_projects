### Data cleaning

```python
df.drop(['veil_type'], axis=1, inplace=True)
```

```python
df['stalk_root'] = df['stalk_root'].apply(lambda x: 'o' if x == '?' else x)
df[df['stalk_root']=='?']
```

```python
for col in df.columns:
    if len(df[col].unique()) == 2:
        print(col,df[col].unique())
```

```python
data = df.copy()
data['class'] = data['class'].apply(lambda x : 1 if x== 'p' else 0)
data['bruises'] = data['bruises'].apply(lambda x : 1 if x== 't' else 0)
data['gill_attachment'] = data['gill_attachment'].apply(lambda x : 1 if x== 'f' else 0)
data['gill_spacing'] = data['gill_spacing'].apply(lambda x : 1 if x== 'c' else 0)
data['gill_size'] = data['gill_size'].apply(lambda x : 1 if x== 'n' else 0)
data['stalk_shape']=data['stalk_shape'].apply(lambda x : 1 if x== 'e' else 0)
data.head()
```

Since ring_number denotes the number of rings 0,1 and 2 initialy(refer attribute information) I am going to revert back to numbers

```python
data['ring_number']=data['ring_number'].apply(lambda x : 0 if x== 'n' else (1 if x=='o' else 2))
```

```python
data['ring_number'].unique()
```

```python
data = pd.get_dummies(data, drop_first=True)
data.head()
```

### Finding the Random state

```python
X = data.drop(['class'], axis = 1)
y = data['class']
```

```python
best_train_score = 0
best_test_score = 0
r_state = 0
for i in range(0, 1000):
    x_train, x_test, y_train, y_test = train_test_split(X,y,test_size = 0.30, random_state=i,stratify=y)
    lr = LogisticRegression()
    lr.fit(x_train, y_train)
    train_score = lr.score(x_train, y_train)
    test_score = lr.score(x_test, y_test)

    #Making sure that the variace is as low as possible
    if round(train_score*100,1) == round(test_score*100,1):
        if test_score > best_test_score:
            best_test_score = test_score
            best_train_score = train_score
            r_state = i
        
print(f'At the Random_State {r_state}, the model performs very well.')
print(f'Train score is {best_train_score*100}')
print(f'Test score is {best_test_score*100}')
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size = 0.30, random_state=1,stratify=y)
```

```python
models = [LogisticRegression(),DecisionTreeClassifier(),KNeighborsClassifier(),GaussianNB(),SVC()]
results = []
m_names = []
for model in models:
    name = model.__class__.__name__
    kfold = KFold(n_splits=10, random_state=1, shuffle=True)
    cv_result = cross_val_score(model, x_train, y_train, cv= kfold, scoring = 'accuracy')
    results.append(cv_result)
    m_names.append(name)
    print(f"{name}: Mean score: {round(cv_result.mean(),3)}  Variance: {round(cv_result.var(),3)}")
```

All models are performing very well. Ill take the top 3 DecisionTreeClassifier, KNN ad SVC for further investigation.

### Feature Importance

```python
from sklearn.tree import DecisionTreeRegressor


model = DecisionTreeRegressor()
model.fit(x_train, y_train)
print(f"Model score: {model.score(x_train, y_train)}")
importance = model.feature_importances_
# summarize feature importance
for i,v in enumerate(importance):
    print('Feature: %s, Score: %.5f' % (x_train.columns[i],v))
# plot feature importance
plt.figure(figsize=(15,8))
plt.bar([x for x in range(len(importance))], importance)
plt.xticks(ticks = np.arange(x_train.shape[1]), labels=x_train.columns, rotation=90)
plt.show()
```

```python
#Picking the top 15 features
df_imp = pd.DataFrame(importance, index = X.columns).sort_values(by = 0, ascending = False).head(15)
```

```python
#Picking features that have importance greater than 0
imp_features = df_imp[df_imp > 0].dropna().index
```

```python
imp_x = data[imp_features]
imp_x.head()
```

```python
#Splitting the important features for training and testing
x_train, x_test, y_train, y_test = train_test_split(imp_x,y,test_size = 0.30, random_state=1,stratify=y)
```

Lets see how the models work with only important features

```python
models = [LogisticRegression(),DecisionTreeClassifier(),KNeighborsClassifier(),GaussianNB(),SVC()]
results = []
m_names = []
for model in models:
    name = model.__class__.__name__
    kfold = KFold(n_splits=10, random_state=1, shuffle=True)
    cv_result = cross_val_score(model, x_train, y_train, cv= kfold, scoring = 'accuracy')
    results.append(cv_result)
    m_names.append(name)
    print(f"{name}: Mean score: {round(cv_result.mean(),3)}  Variance: {round(cv_result.var(),3)}")
```

Again DecisionTree, KNN and SVC are performing well.
Notice that GaussianNB model became too much generalized

I'll choose DecisionTree since the feature importance was derived from it and it gives a very good performance

### Creating final model

Since the model is performing at its best with the default parameters with low bias and low variance for the available data, I am not going for hyperparameter tuning. 

We will need different data to further test the model to see if there are any variance. As for now with the available data the model works well.

```python
#decision tree
dtc = DecisionTreeClassifier()
dtc.fit(x_train,y_train)
y_train_preds = dtc.predict(x_train)
y_test_preds = dtc.predict(x_test)
print("train results:")
print(confusion_matrix(y_train_preds, y_train).ravel())
print(classification_report(y_train_preds, y_train))
print('Test results:')
print(confusion_matrix(y_test_preds, y_test).ravel())
print(classification_report(y_test_preds, y_test))
print(roc_auc_score(y_test, y_test_preds))
```

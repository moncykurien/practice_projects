```python
for col in names:
    plt.figure(figsize=(10,5))
    ax = sns.countplot(x=col,data=df,  palette= ['red','green'])
    total = len(df[col])
    for p in ax.patches:
        percentage = '{:.2f}%'.format(100 * p.get_height()/total)
        x = p.get_x() + p.get_width()/2 - 0.1
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y))
    plt.show()
```

```python
def plot_cat_count(target, cat_var, data):
    plt.figure(figsize=(10,5))
    g = sns.countplot(x=cat_var, hue = target, data=data,palette=['red','green'])
    plt.title(cat_var.title())
    plt.tight_layout()
    plt.legend(['Poisonous','Edible'], loc='upper right')
       
    plt.show()
```

```python
for col in names:
    if col != 'class':
        plot_cat_count('class',col, df)
```
